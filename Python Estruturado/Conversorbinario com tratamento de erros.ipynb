```python
base10 = int(input('Numero na base decimal: '))
i = 2
base2 = []

if base10 == 1:
    base2.insert(0, base10)

while base10 >= i:

    restos = int(base10 % i)
    base10 = int(base10 / i)

    base2.insert(0,restos)

    if base10 == 1:
        base2.insert(0,base10)


bi1 = "".join(map(str,base2))
bi1 = int(bi1)
print(f'Conversão para base binária: {bi1}')
```
