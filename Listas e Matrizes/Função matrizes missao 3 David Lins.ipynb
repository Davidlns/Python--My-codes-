```python
def calc_matrizes(A, B):
  nLinhasA = len(A)
  nColunasA = len(A[0])
  nColunasB = len(B[0])

  M = []

  for linha in range(nLinhasA):
    M.append([])

    for coluna in range(nColunasB):
      M[linha].append(0)

      for k in range(nColunasA):
        M[linha][coluna] += A[linha][k]*B[k][coluna]

  return M
```

```python
##apresentação de resultado e fornecimento de dados

A = [[1,1,1], [2,2,2], [3,3,3]]
B = [[1,1,1], [1,1,1], [1,1,1]]

print(calc_matrizes(A, B))


```
