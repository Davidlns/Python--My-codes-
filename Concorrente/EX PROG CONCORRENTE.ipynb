```python
from time import sleep
from threading import Thread
def tarefa(tempo_espera, mensagem):
    print(f'\iniciando a tarefa {mensagem}')
    sleep(tempo_espera)
    print(f'Conclusão da tarefa {mensagem}')
thread = Thread(target=tarefa, args=(2, 'Thread em execução'))
thread.start()
print('\nAguardando pela execução da Thread...')
thread.join()
print("A execução foi concluída!")
```

```python
from time import sleep
from threading import Thread

def ordem1 (tempo, nome):
  print(f'Executando thread {nome}')
  sleep(tempo)
  print(f'Tarefa {nome} concluida')

x = Thread(target=ordem1, args=(3, 'A'))
y = Thread(target = ordem1, args=(2, 'B'))
x.start()
y.start()
x.join()
y.join()

print("Execução das ordens finalizada com sucesso!")

```

```python
from threading import Thread
from time import sleep

def uma_t (pri, tempo):
  print(f'Thread {pri} iniciada')
  sleep(tempo)
  print(f'Thread {pri} sendo finalizada')

tr1 = Thread(target = uma_t, args = ('A', 3))
tr2 = Thread(target = uma_t, args = ('B', 2))
tr1.start()
tr2.start()
tr1.join()
tr2.join()

print('EXECUÇÃO FINALIZADA!')
```
