```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

class SerieTemporal:
    def __init__(self, dados):
        self.dias = np.array(list(dados.keys())).reshape(-1, 1)
        self.despesas = np.array(list(dados.values()))

        self.modelo = LinearRegression()
        self.modelo.fit(self.dias, self.despesas)

    def visualizar_serie_temporal(self, titulo):
        plt.plot(self.dias[:, 0], self.despesas)
        plt.xlabel("Dias")
        plt.ylabel("Despesas")
        plt.title(f"Série Temporal de {titulo}")
        plt.show()

    def aplicar_regressao_linear(self, titulo):
        previsoes = self.modelo.predict(self.dias)
        plt.plot(self.dias[:, 0], self.despesas, label="Despesas")
        plt.plot(self.dias[:, 0], previsoes, label="Regressão Linear")
        plt.xlabel("Dias")
        plt.ylabel("Despesas")
        plt.title(f"Regressão Linear de {titulo}")
        plt.legend()
        plt.show()

        coeficiente_angular = self.modelo.coef_[0]
        intercepto = self.modelo.intercept_
        print(f"Equação da Regressão Linear: y = {coeficiente_angular:.2f}x + {intercepto:.2f}")

if __name__ == "__main__":
    dados_alimentacao = {1: 100, 2: 120, 3: 150, 4: 180, 5: 200, 6: 220, 7: 250}
    dados_vestuario = {1: 50, 2: 70, 3: 90, 4: 110, 5: 140, 6: 150, 7: 130}
    dados_transporte = {1: 30, 2: 40, 3: 50, 4: 55, 5: 80, 6: 70, 7: 95}

    serie_temporal_alimentacao = SerieTemporal(dados_alimentacao)
    serie_temporal_vestuario = SerieTemporal(dados_vestuario)
    serie_temporal_transporte = SerieTemporal(dados_transporte)

    plt.figure(figsize=(10, 6))

    serie_temporal_alimentacao.visualizar_serie_temporal("Alimentação")
    serie_temporal_alimentacao.aplicar_regressao_linear("Alimentação")

    serie_temporal_vestuario.visualizar_serie_temporal("Vestuário")
    serie_temporal_vestuario.aplicar_regressao_linear("Vestuário")

    serie_temporal_transporte.visualizar_serie_temporal("Transporte")
    serie_temporal_transporte.aplicar_regressao_linear("Transporte")

    plt.xlabel("Dias")
    plt.ylabel("Despesas")
    plt.tight_layout()
    plt.show()
```
