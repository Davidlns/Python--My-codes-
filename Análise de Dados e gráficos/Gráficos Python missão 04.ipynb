```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

dados_alimentacao = {1: 100, 2: 0, 3: 0, 4: 150, 5: 10}
dados_vestuario = {1: 1, 2: 25, 3: 50, 4: 0, 5: 40}
dados_transporte = {1: 0, 2: 0, 3: 100, 4: 300, 5: 50}
```

```python
class SerieTemporal:
    def __init__(self, dados):
        self.dados = dados

    def visualizar_serie_temporal(self, label, cor):
        dias = np.array(list(self.dados.keys()))
        despesas = np.array(list(self.dados.values()))
        plt.plot(dias, despesas, label=label, color=cor, marker='o')

    def aplicar_regressao_linear(self, dados_regressao, cor):
        dias = np.array(list(self.dados.keys())).reshape(-1, 1)
        despesas = np.array(list(self.dados.values()))

        regressao_dias = np.array(list(dados_regressao.keys())).reshape(-1, 1)
        modelo = LinearRegression()
        modelo.fit(dias, despesas)
        previsoes = modelo.predict(regressao_dias)

        plt.plot(regressao_dias, previsoes, label="Regressão Linear", linestyle='dashed', color=cor, marker='x')

serie_alimentacao = SerieTemporal(dados_alimentacao)
serie_vestuario = SerieTemporal(dados_vestuario)
serie_transporte = SerieTemporal(dados_transporte)
```

```python
plt.figure(figsize=(10, 8))

serie_alimentacao.visualizar_serie_temporal("Alimentação", "blue")
serie_vestuario.visualizar_serie_temporal("Vestuário", "green")
serie_transporte.visualizar_serie_temporal("Transporte", "red")

plt.xlabel("Dias")
plt.ylabel("Despesas")
plt.title("Dados de Alimentação, Vestuário e Transporte")

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 6))

serie_alimentacao.visualizar_serie_temporal("Alimentação", "blue")
serie_alimentacao.aplicar_regressao_linear(dados_alimentacao, "blue")
plt.xlabel("Dias")
plt.ylabel("Despesas")
plt.title("Alimentação")

plt.legend()
plt.show()

plt.figure(figsize=(8, 6))

serie_vestuario.visualizar_serie_temporal("Vestuário", "green")
serie_vestuario.aplicar_regressao_linear(dados_vestuario, "green")
plt.xlabel("Dias")
plt.ylabel("Despesas")
plt.title("Vestuário")

plt.legend()
plt.show()

plt.figure(figsize=(8, 6))

serie_transporte.visualizar_serie_temporal("Transporte", "red")
serie_transporte.aplicar_regressao_linear(dados_transporte, "red")
plt.xlabel("Dias")
plt.ylabel("Despesas")
plt.title("Transporte")

plt.legend()
plt.show()
```
