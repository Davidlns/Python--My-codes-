```python
class veiculo ():
  def __init__(self, nome, velocidade_Max, quilometro_litro):

    self.nome = nome
    self.velocidade_Max = velocidade_Max
    self.quilometro_litro = quilometro_litro

  def toStr(self):
    print(f'nome = {self.nome}')
    print(f'velocidade máxima = {self.velocidade_Max}')
    print(f'Quilometros percorridos = {self.quilometro_litro}')

```

```python
class onibus (veiculo):
  pass

  def assentos (self):
    quant = 70
    print(f'A quantidade de assentos é: {quant}')
```

```python
onibus_escolar = onibus("Scania", 120, 10)
onibus_escolar.toStr()
onibus_escolar.assentos()

onibus_escolar2 = onibus_escolar
print(onibus_escolar2)
```
